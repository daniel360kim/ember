```python
a = torch.tensor([2., 3.], requires_grad=True)
b = torch.tensor([6., 4.], requires_grad=True)
```

```python
Q = 3*a**3 - b**2
Q.backward(gradient=torch.tensor([1., 1.]))
```

```python
# check if collected gradients are correct
print(9*a**2 == a.grad)
print(-2*b == b.grad)
```
